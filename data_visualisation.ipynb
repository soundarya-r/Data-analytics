```
from datascience import *
import numpy as np
import pandas as pd
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```
data = pd.read_csv("points.csv", sep = ',')
data
```

```
mean = data.mean()
mean
```

```
mean_data = ["strokes"]
geometric_mean = stats.gmean(data.loc[:,mean_data])
geometric_mean
```

```
harmonic_mean = stats.hmean(data.loc[:,mean_data])
harmonic_mean
```

```
describe_data = data.describe()
describe_data
```

```
first_quartile = describe_data.iloc[4]
third_quartile = describe_data.iloc[6]
inter_quartile = third_quartile - first_quartile
inter_quartile
```
